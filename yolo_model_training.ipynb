```python
# ------------------------------------------
# 🚀 Install Required Packages
# ------------------------------------------
!pip install ultralytics opencv-python


# Importing Required Libraries

from ultralytics import YOLO
import cv2
import os
from collections import defaultdict
from google.colab import files
from google.colab.patches import cv2_imshow

# Toll Pricing Dictionary

vehicle_prices = {
    "car": 60,
    "tempo traveller": 90,
    "bus": 120,
    "truck": 140
}

# Fine tuned YOLOv8 Model

model = YOLO("/content/fine_tuned_yolov8.pt")


def process_video(video_path, output_path="output_video.mp4"):
    vehicle_counter = defaultdict(int)
    vehicle_revenue = defaultdict(int)
    total_toll = 0

    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"❌ Error opening video file: {video_path}")
        return None

    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

    print(f"📹 Processing: {video_path}")
    print(f"📏 Resolution: {frame_width}x{frame_height} | 🎞️ FPS: {fps:.2f}")

    frame_count = 0
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        frame_count += 1
        print(f"\r🔄 Processing frame {frame_count}/{total_frames}", end="")

        results = model(frame)

        for box in results[0].boxes:
            cls_id = int(box.cls[0])
            conf = float(box.conf[0])
            vehicle_type = model.names[cls_id].lower()

            # Updating counters
            vehicle_counter[vehicle_type] += 1
            if vehicle_type in vehicle_prices:
                vehicle_revenue[vehicle_type] += vehicle_prices[vehicle_type]
                total_toll += vehicle_prices[vehicle_type]

            # box coordinates
            x1, y1, x2, y2 = map(int, box.xyxy[0])

            # Custom annotation with toll price
            label = f"{vehicle_type} Toll: ₹{vehicle_prices.get(vehicle_type, 'N/A')} {conf:.2f}"

            # bounding box
            color = (0, 255, 0)  # Green
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)

            # Text background
            (w, h), _ = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 0.6, 1)
            cv2.rectangle(frame, (x1, y1 - h - 10), (x1 + w, y1), color, -1)

            # Text label
            cv2.putText(frame, label, (x1, y1 - 10),
                      cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 0), 1)

        out.write(frame)

    cap.release()
    out.release()

    # detailed report
    print("\n\n📊 Detailed Toll Revenue Report:")
    print("="*75)
    print("{:<20} {:<10} {:<15} {:<20}".format(
        "Vehicle Type",
        "Count",
        "Price/Unit",
        "Total Revenue"))
    print("-"*75)

    for vehicle in sorted(vehicle_counter.keys()):
        count = vehicle_counter[vehicle]
        unit_price = vehicle_prices.get(vehicle, 0)
        revenue = vehicle_revenue.get(vehicle, 0)
        print("{:<20} {:<10} ₹{:<14} ₹{:<19}".format(
            vehicle.title(),
            count,
            unit_price,
            revenue
        ))

    print("="*75)
    print(f"🚗 TOTAL VEHICLES: {sum(vehicle_counter.values())}")
    print(f"💰 TOTAL TOLL REVENUE: ₹{total_toll}")
    print("="*75)

    return output_path

# Uploading video file
print("/content/Video 10s.mp4")
uploaded = files.upload()
video_filename = next(iter(uploaded))

# Processing video
output_file = process_video(video_filename)

#  results
if output_file:
    print("\n🎬 Sample processed frames:")
    cap = cv2.VideoCapture(output_file)
    for _ in range(5):
        ret, frame = cap.read()
        if ret:
            cv2_imshow(frame)
    cap.release()
    files.download(output_file)
else:
    print("❌ Video processing failed")
```
