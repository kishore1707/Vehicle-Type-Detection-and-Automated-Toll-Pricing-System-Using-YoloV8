```python
# ------------------------------------------
# ğŸš€ Install Required Packages (if not already installed)
# ------------------------------------------
# !pip install ultralytics opencv-python

# ------------------------------------------
# ğŸ“¦ Import Required Libraries
# ------------------------------------------
!pip install ultralytics
from ultralytics import YOLO
import cv2
import os

# ------------------------------------------
# ğŸ’° Toll Pricing Dictionary
# ------------------------------------------
vehicle_prices = {
    "car": 60,
    "tempo traveller": 90,
    "bus": 120,
    "truck": 140
}

# ------------------------------------------
# ğŸ§  Load Fine-Tuned YOLOv8 Model
# ------------------------------------------
# NOTE: Update the model path to wherever it's stored
trained_model_path = r"fine_tuned_yolov8.pt"  # If uploaded to Colab, keep it like this
model = YOLO(trained_model_path)

# ------------------------------------------
# ğŸ§ª Detection + Pricing + Display Function
# ------------------------------------------
def detect_vehicle_and_show_price(image_path):
    if not os.path.isfile(image_path):
        print(f"âŒ Image not found: {image_path}")
        return

    # Load image using OpenCV
    image = cv2.imread(image_path)

    # Run detection using YOLOv8
    results = model(image_path)

    detected = False
    for result in results:
        boxes = result.boxes
        class_names = result.names  # Maps class IDs to names

        for box in boxes:
            cls_id = int(box.cls[0])  # class index
            conf = float(box.conf[0])  # confidence
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # bounding box

            vehicle_type = class_names[cls_id]
            price = vehicle_prices.get(vehicle_type.lower(), "Unknown")

            # Draw bounding box and label
            label = f"{vehicle_type} â‚¹{price}"
            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(image, label, (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

            print(f"âœ… Detected: {vehicle_type} | Confidence: {conf:.2f} | ğŸ’° Toll: â‚¹{price}")
            detected = True

    if not detected:
        print("âš ï¸ No vehicles detected.")

    # Display result using OpenCV
    from google.colab.patches import cv2_imshow
    cv2_imshow(image)


# ------------------------------------------
# ğŸ–¼ï¸ Test with Your Image
# ------------------------------------------
# Replace with your test image path (upload to Colab or use local path in local setup)
test_image_path = r"/content/001-o2515_jpg.rf.bac31ad8053e7716b39c0ff38e393693.jpg" # or r"C:\Users\kisho\Downloads\bus-28-.jpg" if running locally
print("\nğŸ” Running Detection:")
detect_vehicle_and_show_price(test_image_path)
test_image_path = r"/content/4bt8bw5mpm1ucmtj5bvk1dcs4l84rh83_jpg.rf.f38f9c0cb830e542a950aff0ffdd9ee8.jpg"# or r"C:\Users\kisho\Downloads\bus-28-.jpg" if running locally
print("\nğŸ” Running Detection:")
detect_vehicle_and_show_price(test_image_path)
test_image_path = r"/content/16_jpg.rf.a54587624193e7ee0ed88f88f8beb0ed.jpg"# or r"C:\Users\kisho\Downloads\bus-28-.jpg" if running locally
print("\nğŸ” Running Detection:")
detect_vehicle_and_show_price(test_image_path)
test_image_path = r"/content/1633889570133_Honda-Vezel-Z-Grade-2014-for-sale_jpg.rf.f5da4a670bc874f7f69a73960b10d7ed.jpg"# or r"C:\Users\kisho\Downloads\bus-28-.jpg" if running locally
print("\nğŸ” Running Detection:")
detect_vehicle_and_show_price(test_image_path)

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ğŸ”¹ Path to the YOLOv8 training results CSV
csv_path = "/content/runs/detect/train/results.csv"

# ğŸ”¸ Load CSV
df = pd.read_csv(csv_path)

# ğŸ” View available columns
print("Available columns:", df.columns.tolist())

# ğŸ“Š Plot Box Loss, Objectness Loss, and Class Loss
plt.figure(figsize=(12, 6))
plt.plot(df['      epoch'], df['          box_loss'], label='Box Loss', color='blue')
plt.plot(df['      epoch'], df['      cls_loss'], label='Class Loss', color='red')
plt.plot(df['      epoch'], df['    dfl_loss'], label='DFL Loss', color='green')
plt.title("YOLOv8 Loss Over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()

# ğŸ“Š Plot mAP50 and mAP50-95
plt.figure(figsize=(10, 5))
plt.plot(df['      epoch'], df['     metrics/mAP_0.5'], label='mAP@0.5', color='purple')
plt.plot(df['      epoch'], df[' metrics/mAP_0.5:0.95'], label='mAP@0.5:0.95', color='orange')
plt.title("Mean Average Precision over Epochs")
plt.xlabel("Epoch")
plt.ylabel("mAP")
plt.legend()
plt.grid(True)
plt.show()

# ğŸ“Š Plot Precision and Recall
plt.figure(figsize=(10, 5))
plt.plot(df['      epoch'], df['    metrics/precision(B)'], label='Precision', color='teal')
plt.plot(df['      epoch'], df['       metrics/recall(B)'], label='Recall', color='magenta')
plt.title("Precision and Recall over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Score")
plt.legend()
plt.grid(True)
plt.show()
```
